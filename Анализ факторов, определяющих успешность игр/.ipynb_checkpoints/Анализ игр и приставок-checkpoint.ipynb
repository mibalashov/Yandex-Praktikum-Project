<h3>1. Изучение общей информации о датасете и предобработка данных</h3>

```python
df = pd.read_csv('games.csv')
df.head()
```

```python
df.columns = map(str.lower, df.columns)
df.head(3)
```

```python
df.info()
```

В датафрейме всего 16715 строк, в некоторых колонках есть пропуски (замена пропусков будет произведена на следующем шаге). 
Названия колонок приведем к нижнему регистру для удобства. 
Типы данных колонок различны, их анализ и изменение будет также проведены в следующих шагах. 

```python
print(df[df['name'].isna()])
df['name']=df['name'].fillna('no_name')
```

Два пропуска в столбце name заменим на no_name.

```python
print(df[df['year_of_release'].isna()])
df['year_of_release']=df['year_of_release'].fillna('1000')
df['year_of_release']=df['year_of_release'].astype('int64')
```

Неизвестный год релиза меняем на "нереальное значение" 1000, чтобы в дальнейшем было удобно работать с данными. 
После, меняем тип данных с object на int64, так как год - целочисленное значение. 

```python
print(df[df['genre'].isna()])
df['genre']=df['genre'].fillna('no_genre')
```

Два пропуска в столбце genre соответсвуют пропускам в столбце name, заменим пропуски на no_genre.

```python
print(df['name'].value_counts().count())
df['name'].value_counts()
df[df['name']=='Need for Speed: Most Wanted']
```

Были найдены игры с несколькими годами релиза на одинаковых платформах (например Need for Speed: Most Wanted, X360, 2005/2015), это повторные релизы (перезапуски игры) с обновлениями. 

```python
print(df['platform'].value_counts().count())
df[df['critic_score'].isna()]['platform'].value_counts().count()
print(df['year_of_release'].value_counts().count())
df[df['critic_score'].isna()]['year_of_release'].value_counts().count()
print(df['critic_score'].max())
df['critic_score']=df['critic_score'].fillna(999)
```

Пропуски в столбце critic_score не зависят от платформы игры и от года релиза тоже не зависят. Т.е. понять зависимость пропусков в рейтингах критиков невозможно, вероятно критики оценивают не все игры, или игры не на всех платформах, поэтому заменяем пропуски на "нереальный рейтинг" 999 для дальнейшего удобства работы с данными. 

```python
print(df['user_score'].isna().sum())
print(df[df['user_score'].isna()]['critic_score'].value_counts())
```

Большая часть пропусков (6663 из 6701) в столбце user_score не имеют рейтига от критиков в столбце critic_score (это рейтинг 999, который проставлен вместо пропусков). Это значит, что данные игры не были популярны, и статистика по рейтингам не была собрана для них. 

```python
print(df['user_score'].value_counts())
df.loc[df['user_score']=='tbd','user_score']=99
df['user_score']=df['user_score'].astype('float64')
df['user_score']=df['user_score'].fillna(99)
```

Кроме пропусков в столбце user_score есть нечисловой рейтинг tbd (означает - будет определен, т.е. это означает, что на данный момент рейтинг отсутвует), заменяем tbd и пропуски на "нереальный рейтинг" 99. Теперь рейтинг 99 - все отсутсвующие рейтинги (пропуски и tbd). 

```python
print(df['rating'].isna().sum())
df[df['rating'].isna()]['critic_score'].value_counts()
#df[df['rating'].isna()]['user_score'].value_counts()
```

Все пропуски в столбце rating не имеют рейтинга от пользователей (user_score) и большая часть пропусков rating (6683 из 6766) не имеют рейтига от критиков (critic_score). Это еще раз подтверждает гипотезу, что данные игры имели низкую популярность и статистика рейтингов по ним не определялась. 

```python
df['rating'].value_counts()
df['rating']=df['rating'].fillna('NR')
df['rating'].value_counts()
```

Меняем пропуски в столбце rating на значение NR (нет рейтинга).  

```python
df.info()
```

***Итог обработки пропусков и изменения типа данных*** - все пропуски в таблице обработанны. Все колонки имеют правильный тип. 

```python
df.duplicated().sum()
```

Дубликаты не обнаруженны. 

```python
df['total_sales']=df['na_sales']+df['eu_sales']+df['jp_sales']+df['other_sales']
```

Для дальнейшего анализа был добавлен столбец total_sales - глобальная выручка по всем регионам из датасета. 

<h3>2. Анализ основных показателей успешности видеоигр</h3>

```python
pivot_year=df.pivot_table(index='year_of_release', values='name', aggfunc='count').reset_index()
pivot_year.columns=['year_of_release','count_release']
pivot_year=pivot_year.sort_values(by='year_of_release')
#print(pivot_year)

pivot_platform=df.pivot_table(index='year_of_release', values='platform', aggfunc=pd.Series.nunique).reset_index()
pivot_platform.columns=['year_of_release','unique_platform']
pivot_platform=pivot_platform.sort_values(by='year_of_release')
#print(pivot_platform)

pivot_year[pivot_year['year_of_release']>1000].plot(x='year_of_release', y='count_release',kind='bar',figsize=(16,3), title='Число релизов')
plt.show()
pivot_platform[pivot_platform['year_of_release']>1000].plot(x='year_of_release', y='unique_platform',kind='bar',figsize=(16,3), title='Число уникальных платформ')
plt.show()
pivot_year
```

По количеству релизов и числу уникальных платформ в различные года видно, что активное развитие игр и платформ началось в период 1993 - 1994 году (1993 - 1994 - переходный период: число релизов (1993г. - 62; 1994 - 121), уникальных платформ (1993г. - 5 ;1994 - 10). ***Таким образом, считаем, что данные до 1993г не информативны, также избавляемся от игр, не имеющих даты релиза (на этапе предобработки им был присвоен год 1000), так их достаточно мало (всего 269), и нет возможности распределить эти игры по годам.***

**Актуальный период для анализа общих закономерностей рынка игр** (определение средних сроков жизни и сроков создания новых платформ, проверка гипотез и другое) - с 1993 г. по 2016 г.

**Актуальный период для принятия решений на будущий 2017 год** - период близкий к прогнозному году (три ретроспективных года), с 2014 по 2016гг. 

```python
df=df[df['year_of_release']>=1993]
df_2014_2016=df[df['year_of_release']>2013]
```

```python
pivot_platform=df.pivot_table(index='platform', values='total_sales', aggfunc='sum').reset_index()
pivot_platform=pivot_platform.sort_values(by='total_sales',ascending=False)
pivot_platform.head(20).plot(x='platform', y='total_sales', kind='bar', figsize=(16,3), title='ТОП-20 платформ по совокупной выручке (1994-2016)')
plt.show()
```

ТОП по выручке довольно логичный - ***в первой семерке представлены самые популярные и конкурирующие семейства приставок: PlayStation (PS2, PS3, PS, PS4), Xbox (X360), Nintendo (Wii, DC),*** однако, бросается в глаза превосходство PlayStation - целых четыре приставки в ТОП - 7 по выручке. Интересно, как обстоят дела в последние годы (последние 3 года). 

Также стоит отметить, что первая шестерка сильно обходит остальные платформы по выручке, так как ***во первых это действительно популярные платформы, а во вторых это платформы которые уже прожили и доживают свой срок жизни,*** т.е они приносят выручку более длительное время, чем платформы которые еще на начальном пути. 

```python
pivot_platform_2013=df_2014_2016.pivot_table(index='platform', values='total_sales', aggfunc='sum').reset_index()
pivot_platform_2013=pivot_platform_2013.sort_values(by='total_sales',ascending=False)
pivot_platform_2013.plot(x='platform', y='total_sales', kind='bar', figsize=(16,3), title='ТОП - платформ по совокупной выручке (2014-2016)')
plt.show()
pivot_platform_2013
```

***И в последние три года лидируют все теже семейства приставок, и среди них PlayStation сильно впереди - если рассматривать приставки одних поколений, то в последние 3 года: PS4 обходит XOne и WiiU, PS3 обходит X360 и Wii.***

```python
pivot_platform_year=df.pivot_table(index='year_of_release',columns='platform',values='total_sales', aggfunc='sum').reset_index()
pivot_platform_year=pivot_platform_year.fillna(0)
pivot_platform_year.plot(x='year_of_release', y=['PS2','PS3','PS','PS4'], figsize=(15,5), grid=True, xticks=pivot_platform_year['year_of_release'],title='Семейство платформ PlayStation')
plt.show()
pivot_platform_year.plot(x='year_of_release', y=['X360','XOne','XB'], figsize=(15,5), grid=True, xticks=pivot_platform_year['year_of_release'],title='Семейство платформ Xbox')
plt.show()
pivot_platform_year.plot(x='year_of_release', y=['Wii','WiiU','GC'], figsize=(15,5), grid=True, xticks=pivot_platform_year['year_of_release'],title='Семейство платформ Nintendo')
plt.show()
```

***Для определения времени жизни и времени создания платформ было проанализировано три самых популярных и конкурирующих семейства платформ (платформы внутри семейств приходили друг другу на смену):*** playstation ('PS','PS2','PS3','PS4'), xbox ('XB','X360','XOne'), nintendo ('GC','Wii','WiiU'). Согласно построенным графикам, среднее время жизни платформы - от 7 до 12 лет; среднее время создание новых платформ внутри семейства (время между релизами платформ) - от 4 до 8 лет. Так же, из графиков можно сделать вывод, что у трех самых популярных семейств платформ стоит ожидать скорый выход новых платформ, так как с момента появления последних платформ внутри семейств прошло 4-5 лет. 
Среднее значение жизни платформы по топ-20 платформам будет рассчитано далее. 

```python
count_year_sales=[]
year_2016=[]
top_platform=['PS2','X360','PS3','Wii','DS','PS','PS4','GBA','PSP','3DS','PC','XB','N64','GC','XOne','GB','SNES','WiiU','PSV','SAT']
for i in top_platform:
    dur_liv_plat=pivot_platform_year[pivot_platform_year[i]!=0][i].count()
    y_2016=pivot_platform_year[pivot_platform_year['year_of_release']==2016][i].sum()
    count_year_sales.append(dur_liv_plat)
    year_2016.append(y_2016)
top_platform_sales=pd.DataFrame({'platform':top_platform, 'count_year_sales':count_year_sales, 'year_2016':year_2016})
top_platform_sales['closed_platform']=top_platform_sales['year_2016'].where(top_platform_sales['year_2016']!=0,'True')
top_platform_sales=top_platform_sales[top_platform_sales['closed_platform']=='True']
print('Медиана жизни платформы',top_platform_sales['count_year_sales'].median())
print('Средняя жизнь платформы',top_platform_sales['count_year_sales'].mean())
top_platform_sales.head(20)
```

Из топ 20 платформ по выручке, только 11 уже перестали приносить выручку к 2016 году, по этим платформам и были посчитаны средние показатели жизни платформы: ***Медиана жизни платформы - 8.0 лет; Средняя жизнь платформы - 8.7 лет.***

Чтобы нивелировать влияние времени (в разные годы были различные цены на игры, различный спрос, различный уровень развития отрасли) анализ разброса данных был проведен за близкий период в три года (2014-2016).

Таким образом, можно сделать вывод, что **различные игры приносят совершенно различный доход, и руководствоваться средними показателями довольно сложно.** 

```python
platform=df_2014_2016.pivot_table(index='platform', values='total_sales', aggfunc='sum').reset_index()
platform.boxplot(column='total_sales', showfliers=False)
plt.show()
platform.describe()
```

***Согласно диаграмме разброса и числовым характеристикам данных по всем платформам вместе- продажи по платформам различаются, но разброс намного меньше, чем у продаж по играм.*** Стандартное отклонение (86) всего в 1.18 раза больше среднего (73 млн долл). Межквартильные расстояние близки (25% - 24 млн долл, 50 (медиана) - 46, 75 - 82), при этом медиана намного ближе к среднему, чем при разбивке продаж по играм. 

В этом случае можно сделать вывод, что **различные платформы приносят различный доход, но средние показатели, в общем, довольно неплохо описывают ситуацию на рынке видео игр.**

```python
def boxplot_platform(df,platform):
    for i in platform:
        df[df['platform']==i].boxplot(column='total_sales', showfliers=False)
        plt.title('Разброс выручки от игр внутри платформы {}'.format(i))
        plt.show()
        display(df[df['platform']==i]['total_sales'].describe())
```

```python
platform_for_box=['PS4', 'XOne', 'WiiU', '3DS']
boxplot_platform(df_2014_2016,platform_for_box)
```

***Диаграммы разброса по выручке от игр внутри популярных платформ (отдельно по платформам) имеют схожий профиль*** - разпределения смещены влево, медианы (PS4 - 0.19 млн. долл, XOne - 0.20, WiiU - 0.19, 3DS - 0.09) и стандартные отклонения близки (кроме плаформы 3DS),а средние значения отличаются: Средняя выручка от игры у PS4 - 0.77 млн. долл, XOne - 0.62, WiiU - 0.59, 3DS - 0.41.

***Платформа 3DS имеет более низкие значения медианы и среднего, это объясняется типом платформы - данная платформа портативная,*** а PS4, XOne и WiiU домашние - это еще один фактор который приводит к существенным различиям выручки от игр (что и показала диаграмма разброса, построенная по всем платформам вместе). 

Таким образом, опираясь на медианы выручки от игр внутри популярных платформ одного типа (домашние или портативные), можно отметить, что ***за последние 3 года отдельные игры популярных платформ приносят в среднем одинаковую выручку.*** Отличие между популярными платформами заключается в величине выбросов (отсюда и различие средней выручки), рекордсменом по этому показателю является PS4 - 14.6 млн. долл (max значение выручки от игры) против 7.4 у XOne и 7.1 у WiiU.  

```python
def scatter_cor(df,platform):
    for i in platform:
        score=df[df['platform']==i]
        score[score['critic_score']!=999].plot(kind='scatter',figsize=(16,3), x='critic_score', y='total_sales',label=i, legend=True, color='green', alpha=0.3)
        plt.title('Зависимость глобальной выручки платформы {} от оценки критиков'.format(i))
        plt.show()
        print('Корреляция:',score[score['critic_score']!=999]['total_sales'].corr(score[score['critic_score']!=999]['critic_score']))
        score[score['user_score']!=99].plot(kind='scatter', figsize=(16,3), x='user_score', y='total_sales',label=i, legend=True, alpha=0.3)
        plt.title('Зависимость глобальной выручки платформы {} от оценки пользователей'.format(i))
        plt.show()
        print('Корреляция:',score[score['user_score']!=99]['total_sales'].corr(score[score['user_score']!=99]['user_score']))
```

```python
scatter_cor(df_2014_2016,platform_for_box)
```

```python
def region_sales(df,region):
    print('-------------- Аналитика по {} ---------------'.format(region))
    print()
    data=['platform','genre','rating']
    for i in data:
        pivot_region_i=df.pivot_table(index=i, values=region, aggfunc='sum').sort_values(by=region, ascending=False)
        total_region_i=pivot_region_i[region].sum()
        pivot_region_i['%_total_region']=pivot_region_i[region]/total_region_i
        print()
        display(pivot_region_i.head(5))
        plt.pie(pivot_region_i['%_total_region'].head(5), labels=pivot_region_i.head(5).index, shadow= True,autopct='%1.0f%%',explode=(0.1,0,0,0,0))
        plt.show()
```

<h3>4. Проверка статистических гипотез</h3> 

```python
XOne_data=df_2014_2016[(df_2014_2016['platform']=='XOne') & (df_2014_2016['user_score']!=99)].reset_index()
XOne_data=XOne_data[['platform','name','user_score']]
XOne_data

PC_data=df_2014_2016[(df_2014_2016['platform']=='PC') & (df_2014_2016['user_score']!=99)].reset_index()
PC_data=PC_data[['platform','name','user_score']]
PC_data
```

```python
results_1 = st.ttest_ind(XOne_data['user_score'], PC_data['user_score'], equal_var = False)
print('p-значение: ', results_1.pvalue)
print(XOne_data['user_score'].mean())
print(PC_data['user_score'].mean())
```

**Гипотеза 1  - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

Нулевая гипотеза - средние пользовательские рейтинги платформ Xbox One и PC равны. Альтернативная гипотеза - средние значения не равны. Критический уровень статистической значимости (альфа)  0.05. 
***Средний пользовательский рейтинг не зависит от платформы (p-значение > альфа, т.е. средние значения могут быть равны с высокой вероятностью), не отвергаем нулевую гипотезу.***

Таким образом, считаем, что средний пользовательский рейтинг не зависит от платформы (могут быть одинаковы), а в данном случае ***у PC средний пользовательский рейтинг выше, чем у Xbox One, 7.1 против 6.5***

```python
action_data=df_2014_2016[(df_2014_2016['genre']=='Action') & (df_2014_2016['user_score']!=99)].reset_index()
action_data=action_data[['genre','name','user_score']]
action_data

sports_data=df_2014_2016[(df_2014_2016['genre']=='Sports') & (df_2014_2016['user_score']!=99)].reset_index()
sports_data=sports_data[['genre','name','user_score']]
sports_data
```

```python
results_2 = st.ttest_ind(action_data['user_score'], sports_data['user_score'], equal_var = False)
print('p-значение: ', results_2.pvalue)
print(action_data['user_score'].mean())
print(sports_data['user_score'].mean())
```

**Гипотеза 2 - Средние пользовательские рейтинги жанров Action и Sports разные.**

Нулевая гипотеза - средние пользовательские рейтинги по жанрам Action и Sports равны. Альтернативная гипотеза - средние значения не равны. Критический уровень статистической значимости (альфа) 0.05. 
***Средний пользовательский рейтинг зависит от жанра (p-значение < альфа, т.е.средние значения могут быть равны с минимальной вероятностью), отвергаем нулевую гипотезу, принимаем альтернативную***

Т.е. средние пользовательские рейтинги по жанрам не равны и в данном случае - средний пользовательский рейтинг по жанру Action: 7.06, а по жанру Sports: 6.95.
