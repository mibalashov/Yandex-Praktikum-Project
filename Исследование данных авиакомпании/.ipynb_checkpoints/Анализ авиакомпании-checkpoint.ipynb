# Исследование данных авиакомпании


**Центральная задача** - анализ городов по числу пребывающих рейсов и анализ популярных моделей самолетов.

<h3>Описание данных</h3>


***В распоряжении два датасета, полученные с помощью sql-запросов:***

**aircraft_data**, в нём содержится информация о: 
model — модели самолета
flights_amount — количество рейсов для каждой модели самолетов model в сентябре 2018 года.

**city_data**, в нём содержится информация о:
city — городах
average_flights — среднем количестве рейсов, прибывающих в город city за день в августе 2018 года.

<h3> Этапы исследования </h3>

1. Изучение общей информации о датасетах и предобработка данных.
2. Анализ данных.
3. Выводы.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

<h3>1. Изучение общей информации о датасетах и предобработка данных</h3>

```python
aircraft_data=pd.read_csv('query_1.csv')
```

```python
display(aircraft_data.head(5))
aircraft_data.info()
print('Пропуски',aircraft_data.isna().mean())
print('Дубликаты',aircraft_data.duplicated().sum())
```

***Пропуски в таблице aircraft_data отсутствуют, типы данных верны, дубликатов нет.***

```python
city_data=pd.read_csv('query_3.csv')
```

```python
display(city_data.head(5))
city_data.info()
print('Пропуски',city_data.isna().mean())
print('Дубликаты',city_data.duplicated().sum())
```

***Пропуски в таблице city_data отсутствуют, дубликатов нет, среднее число рейсов (average_flights) переведем в целочисленный формат - int64***

```python
city_data['average_flights']=city_data['average_flights'].astype('int64')
```

<h3>2. Анализ данных</h3>

```python
display(aircraft_data.sort_values(by='flights_amount'))
aircraft_data.sort_values(by='flights_amount').plot(x='model', y='flights_amount', kind='bar', figsize=(16,4),color='green',title='Число полетов от моделей самолетов')
plt.show()
```

Количество вылетов (за сентябрь 2018 г.) больше у самолетов, предназначенных для ближнемагистральных полетов (до 5000 км) с небольшой вместительностью пассажиров (до 100 пассажиров, самолеты: Cessna 208 Caravan, Bombardier CRJ-200, Sukhoi SuperJet-100), самолеты для малой/средней протяженности маршрута (до 5000 км) и средней вместительностью (до 200 пассажиров, самолеты: Airbus A321-200, Boeing 737-300, Airbus A319-100) занимают средние позиции, а самолеты для дальних перелетов (5000 - 15000 км) с большой вместительностью (от 200 и больше пассажиров, Boeing 767-300, Boeing 777-300) расположились в конце списка. Данная ситуация логична для авиакомпании, выполняющей внутренние пассажирские авиаперевозки.

**Таким образом, наиболее популярным самолетами в авиакомпании «F9» являются ближнемагистральные самолеты с вместительностью до 100 человек, при этом лидер - Cessna 208 Caravan - лёгкий одномоторный самолет с вместительностью 13 человек.**

```python
city_data=city_data.sort_values(by='average_flights', ascending=False)
city_data.sort_values(by='average_flights', ascending=True).plot(x='city', y='average_flights',kind='barh',figsize=(14,23),title='Среднее число прибывающих рейсов в город за день')
plt.show()
```

```python
display(city_data.head(10))
city_data.head(10).sort_values(by='average_flights', ascending=True).plot(x='city', y='average_flights',kind='barh', alpha=0.3,figsize=(16,6),
                                                                          title='Среднее число прибывающих рейсов в город за день по ТОП-10 городам', 
                                                                          color=['g','g','g','g','g','g','g','g','r','r'])
plt.show()
```

```python
city_data['city_category']=city_data['average_flights']
city_data.loc[city_data['average_flights']<=5,'city_category']='меньше 5 рейсов'
city_data.loc[(city_data['average_flights']<=10) & (city_data['average_flights']>5),'city_category']='5-10 рейсов'
city_data.loc[city_data['average_flights']>10,'city_category']='больше 10 рейсов'
#city_data.head(100)
```

***Для дальнейшего анализ среднего числа рейсов по городам было создано 3 категории городов:*** "больше 10 рейсов", "5-10 рейсов", "меньше 5 рейсов".

```python
pivot_fl=city_data.pivot_table(index='city_category',values=['city','average_flights'], aggfunc={'city':'count', 'average_flights':'sum'}).reset_index().sort_values(by='city')
pivot_fl.columns=['city_category','sum_flights','count_city']
pivot_fl['%_all_flights']=pivot_fl['sum_flights']/pivot_fl['sum_flights'].sum()
pivot_fl['%_all_city']=pivot_fl['count_city']/pivot_fl['count_city'].sum()
display(pivot_fl)
plt.pie(pivot_fl['%_all_city'],labels=pivot_fl['city_category'],shadow= True, autopct='%1.0f%%', explode=(0.1,0.1,0.1))
plt.title('% городов по категориям городов')
plt.show()
plt.pie(pivot_fl['%_all_flights'],labels=pivot_fl['city_category'],shadow= True, autopct='%1.0f%%', explode=(0.1,0.1,0.1))
plt.title('% рейсов по категориям городов')
plt.show()
```

Число рейсов в день по городам сильно различается, наибольшую долю занимают города с числом рейсов до 5 в день 76% городов (77 городов) и они приносят 35% рейсов (или 186 рейсов) в день, городов с числом рейсов от 5 до 10 в день 19% (или 19 городов) и они приносят 27% рейсов (142 рейса) в день, а городов с числом рейсов в день более 10 всего 5% (5 городов) и они приносят 38% рейсов (199 рейса). 

**Таким образом, можно сделать вывод, что все три категории городов (несмотря на различную численность) приносят примерно равное число рейсов в день, а значит все направления полетов важны для авиакомпании.** 

<h3>3. Основные выводы</h3>

- ***Наиболее популярными самолетами (по числу рейсов в день) у авиакомпании «F9» являются ближнемагистральные самолеты*** (Cessna 208 Caravan, Bombardier CRJ-200, Sukhoi SuperJet-100) с вместительностью до 100 человек, при этом лидер - Cessna 208 Caravan - лёгкий одномоторный самолет с вместительностью 13 человек.
- ***Рейсы на самолетах для дальних перелетов (5000 - 15000 км) с большой вместительностью (от 200 и больше пассажиров, Boeing 767-300, Boeing 777-300) имеют наименьшую популярность,*** что логично для авиакомпании, выполняющей внутренние рейсы.  
- ***Два города предсказуемо опережают остальные по среднему числу рейсов (прилетов) в день от авиакомпании «F9», это Москва и Санкт-Петербург,*** при этом Москва более чем в 4 раза опережает Санкт-Петербург по этому показателю (129 рейсов против 31).
- ***5 городов (Москва, Санкт-Петербург, Новосибирск, Екатеринбург, Красноярск) из 101 приносят 38% рейсов в день*** для авиакомпании «F9».
- ***Все направления полетов важны для авиакомпании,*** даже города с числом рейсов в день меньше 5, так как в совокупности эти города образуют большую группу и приносят около трети (35%) от всех рейсов в день.
