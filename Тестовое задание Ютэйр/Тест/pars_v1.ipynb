**Парсинг данных с сайта www.krasnodar.vybory.izbirkom.ru/region/krasnodar**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Импорт таблицы со списком id страниц необходимых УИК, таблица получена через слияние внутренних json файлов сайта.

id_kr_district=pd.read_excel('id_kr_district.xlsx')
```

```python
id_str=id_kr_district['id'].tolist()
```

```python
# Функция для парсинга составов УИК

def report_uik(id_str):
    
    path = 'http://www.krasnodar.vybory.izbirkom.ru/region/krasnodar?action=ik&vrn='
    
    URL_head = path + '4234018274986'
    req = requests.get(URL_head)
    soup = BeautifulSoup(req.text, 'lxml')
    heading_table = [] 
    for row in soup.find_all('th'):
        heading_table.append(row.text)    
    
    content=[] 
    uik=[]
    adr=[]
    
    for i in id_str:
        URL = path + str(i)
        req = requests.get(URL)
        soup = BeautifulSoup(req.text, 'lxml')

        table = soup.find('div', attrs={"class":"table margtab"})
    
        table_div = soup.find('div', attrs={"class":"center-colm"})
        table_uik_2 = table_div.find('h2').text
        table_adr_2 = table_div.find('span').text
    
        for row in table.find_all('tr')[1:]:
            uik.append([table_uik_2])
            adr.append([table_adr_2])
            
        for row in table.find_all('tr'):
            if not row.find_all('th'): 
                content.append([element.text for element in row.find_all('td')])

    uik_df=pd.DataFrame(uik)
    adr_df=pd.DataFrame(adr)
    report_uik = pd.DataFrame(content, columns=heading_table)
    
    report_uik['УИК']=uik_df
    report_uik['адрес']=adr_df
    
    display(report_uik.head())
    display(report_uik.tail())
    report_uik.to_csv('report_uik.csv')
```

```python
report_uik(id_str)
```
